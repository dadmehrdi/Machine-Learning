<span style="color:red; font-size:28px; font-weight:bold;">Classification Trees vs Regression Trees</span>

---


## Classification Trees

- **Purpose**: Classify instances into distinct categories.
- **Output**: Discrete labels (e.g., "Yes/No", "Type A/B").
- **Split Criteria**: Gini impurity, entropy, or classification error.
- **Decision Process**: Series of questions leading to a class decision.
- **Use-Cases**: Disease diagnosis, customer segmentation, spam detection.

## Regression Trees

- **Purpose**: Predict a continuous outcome.
- **Output**: Continuous values (e.g., price, temperature).
- **Split Criteria**: Mean squared error (MSE) or mean absolute error (MAE).
- **Decision Process**: Series of questions leading to a numerical prediction.
- **Use-Cases**: House price prediction, sales forecasting, age estimation.

## Key Differences

- **Target Variable**: Discrete for classification trees, continuous for regression trees.
- **Purity Criteria**: Gini/entropy for classification, variance reduction for regression.
- **Prediction Method**: Class assignment vs. average value in leaf nodes.
- **Data Handling**: Segregation of classes vs. reduction of variance.

Both can suffer from overfitting; techniques like pruning and setting maximum depth are used to mitigate this.

<span style="color:red; font-size:28px; font-weight:bold;">Gini Impurity vs Entropy vs Information Gain</span>



## Gini Impurity
- **Definition**: Measures the frequency at which any element of the dataset will be mislabeled when it is randomly labeled.
- **Formula**: `1 - Σ (p_i)^2`, where `p_i` is the probability of an item being classified into class i.
- **Characteristics**: Faster to compute, tends to split the largest class.

## Entropy
- **Definition**: Measures the impurity or randomness in the data.
- **Formula**: `-Σ p_i * log2(p_i)`, where `p_i` is the probability of class i.
- **Characteristics**: Requires more computation, provides a balanced split.

## Information Gain
- **Definition**: Reduction in entropy or impurity after a dataset is split on an attribute.
- **Formula**: `Entropy(Parent) - Weighted Sum of Entropy(Children)`.
- **Use**: Used to determine which feature to split on at each step in building a decision tree.

## Key Points
- **Gini vs Entropy**: Gini is quicker to compute and is preferred for larger datasets; entropy provides more balanced trees.
- **Choosing Criterion**: Depends on the nature of the problem and computational efficiency requirements.
- **Role of Information Gain**: Central in decision tree algorithms, it guides the selection of the splitting attribute.


<span style="color:red; font-size:28px; font-weight:bold;"> Pruning in Decision Trees</span>


- **Purpose**: Reduce tree size to prevent overfitting.
- **Types**:
  1. **Pre-Pruning**: Halts tree growth early (e.g., max depth, min samples per leaf).
  2. **Post-Pruning**: Removes nodes from a fully grown tree (cost complexity pruning).
- **Benefits**:
  - Reduces overfitting.
  - Improves model accuracy on unseen data.
  - Enhances interpretability.


<span style="color:red; font-size:28px; font-weight:bold;"> Cross-Validation in Machine Learning</span>

- **Purpose**: Assess a model’s performance on an independent dataset and reduce overfitting.
- **Method**:
  1. **Split data**: Divide data into 'k' subsets (folds).
  2. **Train and Test**: Train on 'k-1' folds, test on the remaining fold.
  3. **Repeat**: Repeat for each fold and average results.
- **Types**:
  1. **k-Fold Cross-Validation**: Data divided into 'k' folds; commonly used.
  2. **Leave-One-Out (LOO)**: Special case where 'k' equals the number of data points.
- **Benefits**:
  - Provides robust estimate of model performance.
  - Utilizes all available data for training and testing.


<span style="color:red; font-size:28px; font-weight:bold;"> Bagging (Bootstrap Aggregating)</span>


- **Purpose**: Improve the stability and accuracy of machine learning algorithms.
- **Method**:
  - Combine multiple models to get an ensemble model.
  - Each model is trained on a random subset of the data with replacement (bootstrap).
- **Key Features**:
  - Reduces variance, helps prevent overfitting.
  - Models are trained independently.
- **Common Use**: Often used with decision trees, known as Random Forests.
- **Benefits**:
  - Enhances model performance.
  - Reduces the risk of overfitting compared to individual models.


<span style="color:red; font-size:28px; font-weight:bold;"> Out-of-Bag (OOB) Error</span>


- **Definition**: An estimation of prediction error for bagging models, like Random Forests.
- **How it Works**:
  - Each model in a bagging ensemble is trained on a different bootstrap sample from the dataset.
  - Some instances may not be included in a bootstrap sample and are "out of bag" for that model.
  - OOB error is calculated using these out-of-bag instances, which act like a validation set.
- **Advantages**:
  - Provides an unbiased estimate of model error without needing a separate validation set.
  - Efficient use of data, as all observations are used for both training and validation.
- **Usage**:
  - Commonly used as a performance metric for Random Forest models.


<span style="color:red; font-size:28px; font-weight:bold;"> Proximity Matrix in Random Forests</span>


- **Definition**: A matrix that measures the similarity between pairs of instances based on their paths through the decision trees in a Random Forest.
- **How it Works**:
  - When two instances end up in the same leaf node of a tree, their proximity is increased.
  - Proximity is averaged over all trees in the forest.
- **Key Features**:
  - Provides insight into data structure and cluster tendencies.
  - Higher proximity values indicate more similarity between instances.
- **Uses**:
  - Identifying anomalies or outliers.
  - Data imputation.
  - Feature selection and dimensionality reduction.
- **Benefits**:
  - Enhances understanding of the data.
  - Can improve model performance by incorporating proximity information.


<span style="color:red; font-size:28px; font-weight:bold;"> Multidimensional Scaling (MDS) Plot</span>


- **Purpose**: To visualize the similarity or dissimilarity of data points in a low-dimensional space.
- **Key Concept**:
  - MDS reduces the dimensionality of data while preserving the distances (or dissimilarities) between points as much as possible.
- **Types**:
  1. **Metric MDS**: Preserves the pairwise distances between points.
  2. **Non-metric MDS**: Preserves the order of distances, focusing on the rank order.
- **Process**:
  - Starts with a matrix of item-item similarities or dissimilarities.
  - Seeks a low-dimensional representation that reflects these similarities/dissimilarities.
- **Applications**:
  - Visualizing complex datasets with many variables.
  - Exploring patterns and relationships in data.
- **Benefits**:
  - Facilitates intuitive understanding of data structure.
  - Useful in exploratory data analysis.

MDS plots are particularly helpful in revealing hidden structures in the data, making them valuable for tasks like market research, behavioral science, and bioinformatics.


<span style="color:red; font-size:28px; font-weight:bold;"> Stumps in Machine Learning</span>


- **Definition**: A stump is a decision tree with only one split, essentially the simplest form of a decision tree.
- **Structure**:
  - Consists of one root node, which splits into two leaf nodes.
- **Use in Ensemble Methods**:
  - Often used as weak learners in boosting algorithms like AdaBoost.
- **Characteristics**:
  - Simple and interpretable.
  - Low predictive power when used alone.
- **Advantages**:
  - Quick to train and use.
  - In ensemble methods, they contribute to a model that is robust to overfitting.
- **Limitation**:
  - Limited complexity, can capture only very basic patterns in the data.

Stumps are fundamental in understanding the building blocks of more complex ensemble methods in machine learning.


<span style="color:red; font-size:28px; font-weight:bold;"> AdaBoost vs Random Forest</span>


## AdaBoost (Adaptive Boosting)
- **Approach**: Combines multiple weak learners (usually stumps) sequentially.
- **Weight Adjustment**: Increases weights of misclassified instances, so subsequent learners focus more on them.
- **Error Reduction**: Focuses on reducing bias and variance by paying more attention to difficult cases.
- **Model Complexity**: Can become complex as it adds learners to correct errors.
- **Speed**: Generally slower due to the sequential nature of boosting.
- **Best For**: Problems where capturing complex patterns is essential, often at the cost of increased risk of overfitting.

## Random Forest
- **Approach**: Creates an ensemble of decision trees, each trained on a random subset of data and features.
- **Decision Making**: Uses majority voting (for classification) or averaging (for regression) across trees.
- **Error Reduction**: Aims to reduce variance without increasing bias by averaging multiple deep trees.
- **Model Complexity**: Each tree is complex, but the ensemble reduces overall complexity.
- **Speed**: Faster and more parallelizable compared to boosting methods.
- **Best For**: Problems that require a robust model with less risk of overfitting.

## Key Differences
- **Learning Method**: AdaBoost builds sequentially, focusing on mistakes; Random Forest builds trees independently.
- **Handling Overfitting**: AdaBoost can overfit on noisy data, whereas Random Forest is generally more robust.
- **Interpretability**: Random Forest models are typically easier to interpret than AdaBoost models.
- **Flexibility**: AdaBoost is sensitive to noisy data and outliers, while Random Forest handles them better.

Both algorithms are powerful, but their effectiveness can vary significantly depending on the nature of the data and the specific problem at hand.



<span style="color:red; font-size:28px; font-weight:bold;"> AdaBoost vs Gradient Boosting</span>


## AdaBoost (Adaptive Boosting)
- **Learning Approach**: Sequentially adds weak learners (often stumps), adjusting weights of training instances based on the previous model's errors.
- **Error Correction**: Increases weights of misclassified instances, forcing subsequent models to focus on them.
- **Output Update**: Adjusts the final model based on the accuracy of individual weak learners.
- **Complexity Control**: Primarily through the number of iterations (weak learners).
- **Sensitivity**: More sensitive to noisy data and outliers.
- **Best Suited For**: Problems where simpler models need to be combined to improve performance.

## Gradient Boosting
- **Learning Approach**: Sequentially adds models, focusing on correcting the residuals (errors) of the previous models.
- **Error Correction**: Each new model predicts the residual errors of the previous models, refining the overall prediction.
- **Output Update**: Gradient descent used to minimize loss function, improving the model iteratively.
- **Complexity Control**: Through learning rate, tree depth, and number of trees.
- **Sensitivity**: Less sensitive to outliers compared to AdaBoost, but can overfit if not properly tuned.
- **Best Suited For**: Wide range of problems, especially where capturing complex patterns in data is crucial.

## Key Differences
- **Focus of Improvement**: AdaBoost changes instance weights, while Gradient Boosting works on residual errors.
- **Model Updating Method**: AdaBoost uses accuracy to update, Gradient Boosting uses gradient descent.
- **Flexibility and Tuning**: Gradient Boosting offers more hyperparameters for fine-tuning (e.g., learning rate, tree depth).
- **Robustness**: Gradient Boosting generally offers greater robustness to outliers and noisy data.

Both AdaBoost and Gradient Boosting are powerful boosting algorithms, but they differ in their approach to building and improving the ensemble model.


```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree

# Load your data
main_data_df = pd.read_feather("/Users/dadmehr/R/main_data.feather")

# Filter for the year 2021 and select relevant columns
data = main_data_df[main_data_df['YEAR'] == 2021]
data = data[['INCTOT', 'SEX', 'EDUC', 'RACE']]

# Convert INCTOT to a binary variable (1 if income is >= 100000, else 0)
data['High_Income'] = (data['INCTOT'] >= 100000).astype(int)

# Selecting relevant features
X = data[['SEX', 'EDUC', 'RACE']]
y = data['High_Income']

# Splitting the data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Building the decision tree classifier
classifier = DecisionTreeClassifier(random_state=42)
classifier.fit(X_train, y_train)

# Predicting and evaluating the model
y_pred = classifier.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy}")

# Visualizing the tree
plt.figure(figsize=(12, 8))
plot_tree(classifier, feature_names=['SEX', 'EDUC', 'RACE'], class_names=['Low', 'High'], filled=True)
plt.show()
```

```python
# Building a simpler decision tree classifier
# Limiting the depth of the tree
classifier = DecisionTreeClassifier(max_depth=2, random_state=42)
classifier.fit(X_train, y_train)

# Predicting and evaluating the model
y_pred = classifier.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy}")

# Visualizing the tree
plt.figure(figsize=(12, 8))
plot_tree(classifier, feature_names=['SEX', 'EDUC', 'RACE'], class_names=['Low', 'High'], filled=True)
plt.show()
```

# Decision Tree Explanation

This decision tree predicts income levels based on the `EDUC` feature. The root node starts with a high Gini impurity, indicating a mixed classification of income levels.

## Root Node
- **Split Feature**: `EDUC <= 2.5`
- **Gini Impurity**: 0.374, suggesting a fair amount of class mixture.
- **Samples**: 2,276,819, the total number in this node.
- **Value**: [Low: 1,710,686, High: 566,133], showing more individuals fall below the high-income threshold.
- **Majority Class**: Low, the tree will predict a low income if no further information is provided.

## Left Child Node
- **Split Feature**: `EDUC <= 1.5`
- **Gini Impurity**: 0.308, slightly purer than the root, indicating a better separation of income levels.
- **Samples**: 438,081, the subset that meets the left split condition.
- **Value**: [Low: 83,443, High: 354,638], a larger proportion of high income individuals.
- **Prediction**: High, suggesting individuals with `EDUC` at or below 1.5 are more likely to have a high income in this model.

## Right Child Node
- **Split Feature**: `EDUC <= 9.0`
- **Gini Impurity**: 0.204, indicating a better separation than the root node.
- **Samples**: 1,838,738, those that do not meet the left condition.
- **Value**: [Low: 1,627,243, High: 211,495], showing a majority of low income individuals.
- **Prediction**: Low, individuals with `EDUC` above 2.5 up to 9 are more likely to have a low income.

The tree's reliance on `EDUC` for all splits suggests its strong predictive power for income levels in this model. It also implies that within this dataset, the level of education is a significant factor in determining income classification.


```python
import pandas as pd
from sklearn.ensemble import AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree

# Load your data
main_data_df = pd.read_feather("/Users/dadmehr/R/main_data.feather")

# Preprocessing as per your previous code
data = main_data_df[main_data_df['YEAR'] == 2021]
data = data[['INCTOT', 'SEX', 'EDUC', 'RACE']]
data['High_Income'] = (data['INCTOT'] >= 100000).astype(int)

# Selecting relevant features
X = data[['SEX', 'EDUC', 'RACE']]
y = data['High_Income']

# Splitting the data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Initialize AdaBoost with 5 decision stumps
ada_classifier = AdaBoostClassifier(
    base_estimator=DecisionTreeClassifier(max_depth=1),
    n_estimators=3,
    random_state=42
)

# Fit AdaBoost to the training data
ada_classifier.fit(X_train, y_train)

# Predict and calculate accuracy
y_pred = ada_classifier.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print(f"AdaBoost Model Accuracy: {accuracy}")

# Visualize the decision trees within AdaBoost
for tree_num, tree in enumerate(ada_classifier.estimators_):
    plt.figure(figsize=(5, 3))
    plot_tree(tree, feature_names=X_train.columns, class_names=['Low', 'High'], filled=True)
    plt.title(f'Decision Tree {tree_num + 1} in the AdaBoost Model')
    plt.show()
```

# AdaBoost Decision Trees Interpretation

AdaBoost selects features to split on based on their ability to reduce the model's error. In the given decision trees, `EDUC` is consistently chosen as the splitting feature, implying its strong relationship with the target variable (income level). This suggests that `EDUC` is a dominant predictor and provides the most information gain for classifying individuals into income brackets.

## Decision Tree 1
- **Primary Split**: Based on `EDUC`.
- **Insight**: Lower levels of education (`EDUC <= 2.5`) lead to a higher probability of a six-digit income in this tree, which might be counterintuitive and warrants further investigation.

## Decision Tree 2
- **Primary Split**: Also on `EDUC`.
- **Insight**: Shows a tendency to predict a six-digit income for individuals with an education level above the ninth grade (`EDUC > 9.0`).

## Decision Tree 3
- **Primary Split**: Once again on `EDUC`.
- **Insight**: Suggests that individuals with the lowest education levels (`EDUC <= 1.5`) are more likely to have a six-digit income according to this tree, which could be due to the adaptive boosting focusing on correcting previous misclassifications.

In all trees, the `EDUC` feature is pivotal for predicting income level. If `EDUC` provides a clear distinction between different income levels, AdaBoost will prioritize it for splits to effectively reduce classification error. This reliance on `EDUC` could be indicative of its predictive power in the context of the dataset, overshadowing other features like `SEX`.


```python
import pandas as pd
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree

# Load your data
main_data_df = pd.read_feather("/Users/dadmehr/R/main_data.feather")

# Preprocessing as per your previous code
data = main_data_df[main_data_df['YEAR'] == 2021]
data = data[['INCTOT', 'SEX', 'EDUC', 'RACE']]
data['High_Income'] = (data['INCTOT'] >= 100000).astype(int)

# Selecting relevant features
X = data[['SEX', 'EDUC', 'RACE']]
y = data['High_Income']

# Splitting the data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Initialize Gradient Boosting Classifier
gbc = GradientBoostingClassifier(n_estimators=100, learning_rate=0.1, max_depth=3, random_state=42)

# Fit Gradient Boosting to the training data
gbc.fit(X_train, y_train)

# Predict and calculate accuracy
y_pred = gbc.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print(f"Gradient Boosting Model Accuracy: {accuracy}")

# Visualize the first tree from the gradient boosting model
# The first index [0] accesses the first tree, and the second index [0] accesses the single decision tree within the list.
first_tree = gbc.estimators_[0, 0]
plt.figure(figsize=(20, 10))
plot_tree(first_tree, feature_names=X_train.columns, class_names=['Low', 'High'], filled=True)
plt.show()
```

# Gradient Boosting Model Tree Interpretation

This decision tree is part of a Gradient Boosting model used to predict whether individuals earn a six-digit income, based on `EDUC`, `SEX`, and `RACE`.

- **Root Split**: `EDUC <= 2.5` indicates education level is a significant predictor.
- **Left Branch**: Further splits by `EDUC` suggest finer distinctions in income levels at lower education levels.
- **Right Branch**: Splits by `EDUC` again and then by `SEX`, pointing to a combined effect of education and gender on income.
- **Leaf Nodes**: Represent the model's final predictions with `value` indicating the likelihood of high income.
- **Model Accuracy**: Achieved an accuracy of 0.8708, indicating a strong predictive performance.

The `friedman_mse` values indicate the error reduction each split brings to the model, guiding the gradient boosting process.
