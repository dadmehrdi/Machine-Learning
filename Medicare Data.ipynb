# Loading Medicare Data from Multiple CSV Files

```python
import pandas as pd


data_13 = pd.read_csv('Medicare Data/MUP_PHY_R19_P04_V10_D13_Geo.csv')
data_14 = pd.read_csv('Medicare Data/MUP_PHY_R19_P04_V10_D14_Geo.csv')
data_15 = pd.read_csv('Medicare Data/MUP_PHY_R19_P04_V10_D15_Geo.csv')
data_16 = pd.read_csv('Medicare Data/MUP_PHY_R19_P04_V10_D16_Geo.csv')
data_17 = pd.read_csv('Medicare Data/MUP_PHY_R19_P04_V10_D17_Geo.csv')
data_18 = pd.read_csv('Medicare Data/MUP_PHY_R20_P04_V10_D18_Geo.csv')
data_19 = pd.read_csv('Medicare Data/MUP_PHY_R21_P04_V10_D19_Geo.csv')
data_20 = pd.read_csv('Medicare Data/MUP_PHY_R22_P05_V10_D20_Geo.csv')
data_21 = pd.read_csv('Medicare Data/MUP_PHY_R23_P05_V10_D21_Geo.csv')
```

```python
print(data_21.head())
```

```python
# Concatenate all the dataframes into one
medicare_data = pd.concat([data_13, data_14, data_15, data_16, data_17, data_18, data_19, data_20, data_21], ignore_index=True)
```

```python
medicare_data
```

# KDE Visualization of Middle 50% Distribution of Total Services in Medicare Data

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Calculate the 5th and 95th percentile
lower_bound = medicare_data['Tot_Srvcs'].quantile(0.25)
upper_bound = medicare_data['Tot_Srvcs'].quantile(0.75)

# Filter the data to retain only the middle 90%
filtered_data = medicare_data[(medicare_data['Tot_Srvcs'] >= lower_bound) & (medicare_data['Tot_Srvcs'] <= upper_bound)]

# Plotting the KDE for the filtered distribution of Total Services
plt.figure(figsize=(10, 6), dpi=400)
sns.kdeplot(filtered_data['Tot_Srvcs'], shade=True, color='skyblue')
plt.title('Distribution of Total Services (Middle 50%)')
plt.xlabel('Total Services')
plt.ylabel('Density')
plt.grid(True, which='both', linestyle='--', linewidth=1)
plt.tight_layout()
plt.show()
```

# Visualization of Top HCPCS Descriptions in Medicare Data

```python
import matplotlib.pyplot as plt

# Calculate the frequencies of each HCPCS_Desc
hcpcs_frequencies = medicare_data['HCPCS_Desc'].value_counts()

# Display the top HCPCS_Desc with their frequencies
top_hcpcs = hcpcs_frequencies.head(15)  # Adjust the number to show more or fewer top values

# Plotting
plt.figure(figsize=(12, 8), dpi=400)
top_hcpcs.plot(kind='barh', color='skyblue')
plt.title('Top HCPCS Descriptions and Their Frequencies')
plt.xlabel('Frequency')
plt.ylabel('HCPCS Description')
plt.gca().invert_yaxis()  # To display the highest frequency at the top
plt.tight_layout()
plt.show()
```

# Visualizing the Top 15 Costliest Medical Services Using Truncated HCPCS Descriptions

```python
# Truncating the HCPCS_Desc to the first 50 characters and appending '...'
medicare_data['Truncated_Desc'] = medicare_data['HCPCS_Desc'].apply(lambda x: x[:60] + '...' if len(x) > 50 else x)

# Grouping by the truncated description and calculating the average cost
average_costs = medicare_data.groupby('Truncated_Desc')['Avg_Sbmtd_Chrg'].mean().sort_values(ascending=False)

# Getting the top ten costliest reasons
top_ten_costliest = average_costs.head(15)

# Plotting
plt.figure(figsize=(12, 8), dpi=400)
top_ten_costliest.plot(kind='barh', color='skyblue')
plt.title('Top 15 Costliest Reasons Based on HCPCS Descriptions')
plt.xlabel('Average Submitted Charge')
plt.ylabel('Truncated HCPCS Description')
plt.gca().invert_yaxis()  # To display the costliest reason at the top
plt.tight_layout()
plt.show()
```
